5. Write a sorting algorithm for a list in Python

```
def sort(mylist):
  n=len(mylist)
  for i in range(n):
    for j in range(n-i-1):
      if mylist[j] > mylist[j+1]:
        mylist[j], mylist[j+1] = mylist[j+1], mylist[j]
  print(mylist)
sort([80, 55, 70, 0, -10])
```

6. Write a function that returns elements in a list without duplicating.

```
def uniquelist(mylist):
  unique=[]
  for x in mylist:
    if x not in unique:
      unique.append(x)
  return unique
uniquelist([1,2,5,6,8,9,0,8,1,2,8,1])
```

7. Write a function that returns the repeated elements in a list

```
def duplicatecheck(mylist):
  mylist.sort()
  duplicate=[]
  for i in range(len(mylist)-1):
    if mylist[i]==mylist[i+1]:
      if mylist[i] not in duplicate:
       duplicate.append(mylist[i])
  return duplicate
duplicatecheck([2,3,2,1,2,2,5,6,9,-1,10,5,4,7])
```

8. Write a function that calculates the maximum difference between two elements in an array such that larger element appears after the smaller number.

```
def maxdiff(myarray):
  current_max=myarray[1]-myarray[0]
  for i in range(0, len(myarray)):
    for j in range(i+1, len(myarray)):
      if (myarray[j]-myarray[i])>current_max:
        current_max=(myarray[j]-myarray[i])
  return current_max
maxdiff([1,-2,10,-2,6])
      
```
